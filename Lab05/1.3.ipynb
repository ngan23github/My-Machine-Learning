## Tính mean, median, mode, variance, standard deviation, range, percentile, quartile, interquartile range (IQR) (alcohol)

```python
# Mean
wine_mean = np.mean(wine["alcohol"])
print("Mean:", wine_mean)

# Median
wine_median = np.median(wine["alcohol"])
print("Median:", wine_median)

# Mode
wine_mode = stats.mode(wine["alcohol"])
print("Mode:", wine_mode)

# Variance
wine_variance = np.var(wine["alcohol"])
print("Variance:", wine_variance)

# Standard Deviation
wine_sd =  np.std(wine["alcohol"])
print("Std:", wine_sd)

# Max
wine_max = np.max(wine["alcohol"])
print("Max:", wine_max)

# Min
wine_min = np.min(wine["alcohol"])
print("Min:", wine_min)

# Range
wine_range = wine_max - wine_min
print("Range:", wine_range)

# Percentile 
wine_percentile_60 = np.percentile(wine["alcohol"], 60)
print("60th percentile:", wine_percentile_60)

# Quartile
wine_quartile = np.quantile(wine["alcohol"],0.75)
print("Quartile:", wine_quartile)

# IQR
wine_IQR = stats.iqr(wine["alcohol"])
print("IQR:", wine_IQR)
```

- Mean: độ cồn trung bình của các mẫu rượu.

- Median: nửa số chai có độ cồn thấp hơn, nửa còn lại cao hơn.

- Mode: mức độ cồn xuất hiện nhiều nhất.

- Variance: độ chênh lệch về độ cồn giữa các mẫu.

- Standard deviation: trung bình mỗi chai lệch bao nhiêu đơn vị cồn so với mean.

- Min/Max: độ cồn thấp nhất và cao nhất.

- Range: khoảng chênh giữa chai có độ cồn cao nhất và thấp nhất.

- Percentile (60%): 60% số chai có độ cồn ≤ giá trị này.

- Quartiles (Q1, Q2, Q3): chia dữ liệu thành 4 nhóm theo độ cồn.

- IQR (Q3 – Q1): độ phân tán của 50% mẫu rượu nằm giữa.
