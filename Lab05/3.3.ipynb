# Sử dụng SweetViz trên data Marketing Campaign

## Import thư viện

```python
import pandas as pd
import sweetviz as sv
```

## Đọc dữ liệu

```python
data = pd.read_csv("data/marketing_campaign.csv", sep="\t")
```

## Xử lý

```python
# Tạo báo cáo EDA tự động
report = sv.analyze(data)

# Xuất báo cáo thành file HTML
report.show_html('marketing_campaign_sweetviz_report.html')
```
