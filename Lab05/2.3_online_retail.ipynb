# ONLINE RETAIL - Trực quan hóa dữ liệu

## Import thư viện

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Đọc dữ liệu

```python
retail = pd.read_excel("./data/Online Retail.xlsx")
```

## Trực quan hóa dữ liệu

```python
# Tính doanh thu
retail['Revenue'] = retail['Quantity'] * retail['UnitPrice']

# Top 10 sản phẩm bán nhiều nhất
top_products = retail['Description'].value_counts().head(10)
```

### Case 1: Basic

```python
plt.figure(figsize=(12,6))
plt.bar(top_products.index, top_products.values)
plt.xticks(rotation=75)
plt.show()
```

### Case 2: Advanced 1

```python
plt.figure(figsize=(12,6))
plt.bar(top_products.index, top_products.values)
plt.title("Top 10 sản phẩm bán chạy nhất", fontsize=15)
plt.xlabel("Sản phẩm", fontsize=12)
plt.ylabel("Số lượng bán", fontsize=12)
plt.xticks(rotation=75)
plt.show()
```

### Case 3: Advanced 2

```python
top_revenue = retail.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)

fig, ax = plt.subplots(1,2, figsize=(20,8))
ax[0].bar(top_products.index, top_products.values)
ax[0].set_title("Top 10 sản phẩm bán theo số lượng")
ax[0].tick_params(axis='x', rotation=75)

ax[1].bar(top_revenue.index, top_revenue.values)
ax[1].set_title("Top 10 sản phẩm theo doanh thu")
ax[1].tick_params(axis='x', rotation=75)
plt.show()
```
