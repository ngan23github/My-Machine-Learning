# RED WINE QUALITY - Trực quan hóa dữ liệu

## Import thư viện

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Đọc dữ liệu

```python
wine = pd.read_csv("./data/winequality-red.csv")
wine.head()
```

## Trực quan hóa dữ liệu

### Sắp xếp theo Alcohol

```python
wine_sorted = wine.sort_values('alcohol', ascending=False)
```

### Case 1: Basic

```python
plt.figure(figsize=(12,6))
x = wine_sorted['quality'].value_counts().index
y = wine_sorted['quality'].value_counts().values
plt.bar(x,y)
plt.show()
```

### Case 2: Advanced 1

```python
# plt.figure(figsize=(12,6))
# plt.bar(x,y)
plt.title('Số lượng mẫu theo Quality', fontsize=15)
# plt.xlabel('Quality', fontsize=12)
# plt.ylabel('Số lượng mẫu', fontsize=12)
sns.barplot(x=wine['quality'], y=wine['alcohol'], palette='Oranges')
plt.show()
```

### Case 3: Advanced 2

```python
fig, ax = plt.subplots(figsize=(16,6), ncols=2)
sns.barplot(x=wine['quality'], y=wine['alcohol'], ax=ax[0], estimator='mean', palette='Blues')
ax[0].set_title("Alcohol trung bình theo Quality")
sns.barplot(x=wine['quality'], y=wine['sulphates'], ax=ax[1], estimator='mean', palette='Greens')
ax[1].set_title("Sulphates trung bình theo Quality")
plt.show()
```
