# PIMA INDIANS DIABETES - Trực quan hóa dữ liệu

## Import thư viện

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Đọc dữ liệu

```python
diabetes = pd.read_csv("./data/diabetes.csv")
```

## Trực quan hóa dữ liệu

### Case 1: Basic

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Outcome', data=diabetes)
plt.show()
```

### Case 2: Advanced 1

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Outcome', data=diabetes)
ax.set_title("Số lượng bệnh nhân theo Outcome", fontsize=15)
ax.set_xlabel("Outcome (0 = Không tiểu đường, 1 = Có tiểu đường)")
ax.set_ylabel("Số lượng")
plt.show()
```

### Case 3: Advanced 2

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))
sns.barplot(x='Outcome', y='Glucose', data=diabetes, ax=ax[0], estimator='mean')
ax[0].set_title("Glucose trung bình theo Outcome")
sns.barplot(x='Outcome', y='BMI', data=diabetes, ax=ax[1], estimator='mean')
ax[1].set_title("BMI trung bình theo Outcome")
plt.show()
```
