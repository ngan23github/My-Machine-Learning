# Sử dụng AutoViz trên data Marketing Campaign

## Import thư viện

```python
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class
```

## Xử lý

```python
# Khởi tạo AutoViz
AV = AutoViz_Class()

# Đọc file CSV (file Kaggle Marketing Campaign thường phân cách bằng ;)
df = pd.read_csv("data/marketing_campaign.csv", sep="\t", encoding="utf-8")

# Kiểm tra nhanh
print(df.shape)
print(df.columns[:10])  # in thử 10 cột đầu
```

```python
df_av = AV.AutoViz(
    filename="",         
    sep="\t",            
    depVar="Income",     # hoặc None nếu chỉ khám phá chung
    dfte=df,             
    header=0,
    verbose=1,
    lowess=False
)
```
